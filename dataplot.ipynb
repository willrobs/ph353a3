```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
from math import *
```

```python
mu=10.0
m=2.0
a=0.05
niter=10000
points=100
smu=mu,
sm=m,
sma=a,
sniter=niter,
spoints=points,
smu=str(smu)
sm=str(sm)
sma=str(sma)
sniter=str(sniter)
spoints=str(spoints)
name=smu+sm+sma+sniter+spoints
```

# Varying elastic constant

```python
gse=[] ####ground state energies to plot against mu
fee=[] ####first excited state energies ""
gse_err=[]
fee_err=[]

for h in range(2,12,2):
    h=h+0.0
    smu=h,
    smu=str(smu) ####for calling respective file
    name=smu+sm+sma+sniter+spoints
    
    data=pandas.read_csv('info'+name+'.txt', delim_whitespace=True,skiprows=1)
    
    gse=np.append(gse, data.col1[0]) 
    gse_err=np.append(gse_err, data.col2[0])
    fee=np.append(fee, data.col1[1] + data.col1[0])
    fee_err=np.append(fee_err, data.col2[1] + data.col2[0])

mu=np.arange(5, 45, 5)

plt.figure(1)
plt.title('Ground State Energy')
plt.xlabel('Elastic constant')
plt.ylabel('Energy')
plt.errorbar(mu,gse,yerr=gse_err,fmt='.')
plt.plot(mu, mu/((4*m)**0.5),'r')
plt.show()

plt.figure(2)
plt.title('First Excited State Energy')
plt.xlabel('Elastic constant')
plt.ylabel('Energy')
plt.errorbar(mu,fee,yerr=fee_err,fmt='.')
plt.plot(mu, (3*mu)/((4*m)**0.5),'r')
plt.show()

gse_re=[]
fee_re=[]
gse_err_re=[]
fee_err_re=[]

#mups=[7.0,9.0,12.0,14.0,17.0,19.0,22.0,27.0,32.0,37.0,42.0]
mups=[7.0,9.0,12.0,14.0,17.0,22.0,27.0,32.0,37.0,42.0]
for h in mups:
    smu=h,
    smu=str(smu) ####for calling respective file
    name=smu+sm+sma+sniter+spoints
    
    data=pandas.read_csv('info'+name+'.txt', delim_whitespace=True)
    
    gse_re=np.append(gse_re, data.col1[0]) 
    gse_err_re=np.append(gse_err_re, data.col2[0])
    fee_re=np.append(fee_re, data.col1[1] + data.col1[0])
    fee_err_re=np.append(fee_err_re, data.col2[1] + data.col2[0])

#mu=[5,7,9,10,12,14,15,17,19,20,22,25,27,30,32,35,37,40,42]

plt.figure(3)
plt.title('Ground State Energy')
plt.xlabel('Elastic constant')
plt.ylabel('Energy')
plt.errorbar(mu,gse,yerr=gse_err,fmt='.')
plt.errorbar(mups,gse_re,yerr=gse_err_re,fmt='.')
plt.plot(mu, mu/((4*m)**0.5))
plt.show()

plt.figure(4)
plt.title('First Excited State Energy')
plt.xlabel('Elastic constant')
plt.ylabel('Energy')
print(len(fee_re))
print(len(fee_err))
plt.errorbar(mu,fee,yerr=fee_err,fmt='.')
plt.errorbar(mups,fee_re,yerr=fee_err_re,fmt='.')
plt.plot(mu, (3*mu)/((4*m)**0.5))
plt.show()
    
```

# Varying m

```python
gse=[] ####ground state energies to plot against mu
fee=[] ####first excited state energies ""
gse_err=[]
fee_err=[]

for h in range(2,12,2):
    h=h+0.0
    sm=h,
    sm=str(sm) ####for calling respective file
    name=smu+sm+sma+sniter+spoints
    
    data=pandas.read_csv('info'+name+'.txt', delim_whitespace=True,skiprows=1)
    
    gse=np.append(gse, data.col1[0]) 
    gse_err=np.append(gse_err, data.col2[0])
    fee=np.append(fee, data.col1[1] + data.col1[0])
    fee_err=np.append(fee_err, data.col2[1] + data.col2[0])

m_array=np.arange(2, 12, 2)
x_axis=np.arange(0,15,0.1)

plt.figure(1)
plt.title('Ground State Energy')
plt.xlabel('Mass')
plt.ylabel('Energy')
plt.errorbar(m_array,gse,yerr=gse_err,fmt='.')
plt.plot(x_axis, mu/((4*x_axis)**0.5),'r')
plt.axis([1,11,1,max(gse) +1])
plt.show()

plt.figure(2)
plt.title('First Excited State Energy')
plt.xlabel('Mass')
plt.ylabel('Energy')
plt.errorbar(m_array,fee,yerr=fee_err,fmt='.')
plt.plot(x_axis, (3*mu)/((4*x_axis)**0.5),'r')
plt.axis([1,11,4,max(fee) +1])
plt.show()

gse_re=[]
fee_re=[]
gse_err_re=[]
fee_err_re=[]

for h in range(1,13,2):
    h=h+0.0
    sm=h,
    sm=str(sm) ####for calling respective file
    name=smu+sm+sma+sniter+spoints
    
    data=pandas.read_csv('info'+name+'.txt', delim_whitespace=True)
    
    gse_re=np.append(gse_re, data.col1[0]) 
    gse_err_re=np.append(gse_err_re, data.col2[0])
    fee_re=np.append(fee_re, data.col1[1] + data.col1[0])
    fee_err_re=np.append(fee_err_re, data.col2[1] + data.col2[0])
mp_array=np.arange(1,13,2)    
plt.figure(3)
plt.title('Ground State Energy')
plt.xlabel('Mass')
plt.ylabel('Energy')
plt.errorbar(m_array,gse,yerr=gse_err,fmt='.')
plt.errorbar(mp_array,gse_re,yerr=gse_err_re,fmt='.')
plt.plot(x_axis, mu/((4*x_axis)**0.5))
plt.show()

plt.figure(4)
plt.title('First Excited State Energy')
plt.xlabel('Mass')
plt.ylabel('Energy')
plt.errorbar(m_array,fee,yerr=fee_err,fmt='.')
plt.errorbar(mp_array,fee_re,yerr=fee_err_re,fmt='.')
plt.plot(x_axis, (3*mu)/((4*x_axis)**0.5))
plt.show()
    
```
